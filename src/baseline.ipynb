```python
import pandas as pd
from numpy import random
randomState = 42
random.seed(randomState)
```

```python
def aleatoire(text):
  possibilities = ["Entrée", "Plat principal", "Dessert"]
  r = random.randint(3)
  return possibilities[r]


def majoritaire(text):
  return "Plat principal"
```

```python
train_df = pd.read_csv("../data/train.csv")
test_df = pd.read_csv("../data/test.csv")

cpt_aleatoire = 0
for index, row in test_df.iterrows() :
    res = aleatoire(row["titre"])
    if res == row["type"] :
        cpt_aleatoire += 1

print("Résultat aléatoire : " + str(cpt_aleatoire/len(test_df)))

cpt_maj = 0
for index, row in test_df.iterrows() :
    res = majoritaire(row["titre"])
    if res == row["type"] :
        cpt_maj += 1

print("Résultat majoritaire : " + str(cpt_maj/len(test_df)))
```
