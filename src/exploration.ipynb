```python
import pandas as pd
import seaborn as sns
import spacy
```

```python
train_df = pd.read_csv("../data/train.csv")
test_df = pd.read_csv("../data/test.csv")
print(train_df.head())
```

```python
print("Nombre de documents d'entraînement : ", len(train_df))
print("Nombre de documents de test : ", len(test_df))
```

```python
res = {}

for _ , doc in train_df.iterrows():
    type = doc["type"]

    if type not in res:
        res[type] = 1
    else : 
        res[type] += 1

for key in sorted(res, key=res.get, reverse=True):
    print(f"{key} : {res[key]}")
```

 Il y a plus de plats principaux que de desserts et d'entrées. Environ autant de plats que d'entrées et de dessert réunis.

```python
sDessert = 0
sEntrée = 0
sPlat = 0
for row, rec in train_df.iterrows():
    if rec["type"] == "Dessert":
        sDessert += len(rec["recette"])
    elif rec["type"] == "Entrée":
        sEntrée += len(rec["recette"])
    elif rec["type"] == "Plat principal":
        sPlat += len(rec["recette"])

sTotal = sDessert + sEntrée + sPlat

print("Longueur moyenne des recettes : ", sTotal / len(train_df), "caractères")
print("Longueur moyenne des recettes de dessert : ", sDessert / res["Dessert"], "caractères")
print("Longueur moyenne des recettes d'entrée : ", sEntrée / res["Entrée"], "caractères")
print("Longueur moyenne des recettes de plat : ", sPlat / res["Plat principal"], "caractères")
```

```python
sDessert = 0
sEntrée = 0
sPlat = 0
for row, rec in train_df.iterrows():
    if rec["type"] == "Dessert":
        sDessert += len(rec["ingredients"].split(" - "))
    elif rec["type"] == "Entrée":
        sEntrée += len(rec["ingredients"].split(" - "))
    elif rec["type"] == "Plat principal":
        sPlat += len(rec["ingredients"].split(" - "))

sTotal = sDessert + sEntrée + sPlat

print("Nombre d'ingrédients différents : ", sTotal / len(train_df), "caractères")
print("Nombre d'ingrédients différents dans les recettes de dessert : ", sDessert / res["Dessert"], "caractères")
print("Nombre d'ingrédients différents dans les recettes d'entrée : ", sEntrée / res["Entrée"], "caractères")
print("Nombre d'ingrédients différents dans les recettes de plat : ", sPlat / res["Plat principal"], "caractères")
```

```python
stop_words_cuisine = { # Générés par Gemini
    # --- Unités et Mesures ---
    "g", "gr", "gramme", "grammes", "kg", "kilogramme",
    "l", "litres", "litre", "ml", "cl", "dl",
    "cuillère", "cuillères", "c.à.s", "c.à.c", "cas", "cac",
    "pincée", "pincées", "tasse", "tasses", "verre", "verres",
    "poignée", "tranche", "tranches", "morceau", "morceaux",
    "bouteille", "sachet", "boîte", "pot", "paquet", "brin", 
    "feuille", "feuilles", "gousse", "gousses", "°","c", "mn",
    
    # --- Verbes d'action (Instructions fréquentes) ---
    "ajouter", "ajoutez", "mélanger", "mélangez", "remuer",
    "verser", "versez", "incorporer", "couper", "coupez",
    "hacher", "éplucher", "cuire", "laisser", "mettre", "mettez",
    "servir", "disposer", "saupoudrer", "garnir", "préparer",
    "chauffer", "préchauffer", "bouillir", "mijoter", "réserver",
    "laver", "égoutter", "battre", "fouetter", "étaler", "faire",
    
    # --- Adjectifs / États ---
    "chaud", "froide", "tiède", "bouillant", 
    "fondu", "haché", "râpé", "entier", "entière",
    "fin", "fine", "épais", "gros", "petit",
    "frais", "fraîche", "sec", "sèche",
    
    # --- Ingrédients "neutres" (souvent exclus de l'analyse thématique) ---
    # À retirer de cette liste si vous voulez garder tous les ingrédients
    "sel", "poivre", "eau", "huile", "beurre", "sucre", "farine",
    
    # --- Mots de liaison temporels ---
    "puis", "ensuite", "enfin", "pendant", "environ", "minutes", "min", 
    "heure", "heures", "seconde", "secondes", "avant", "après",

    # --- Autres ---
    " ", "1/2","bien"
}

def tokenize(texts, words, nlp):
  for text in texts:
    for token in nlp(text):
      if not (token.is_stop or token.is_punct or token.text.lower() in stop_words_cuisine or token.text.isdigit()):
        if token.text not in words:
          words[token.text] = 1
        else:
            words[token.text] += 1
  return words

def tokenize_ingredients_recettes(df, nlp):
  words = tokenize(df["recette"], {}, nlp)
  words = tokenize(df["ingredients"], words, nlp)
  return words


nlp = spacy.load("fr_core_news_sm")
plats_df = train_df[train_df["type"] == "Plat principal"]
desserts_df = train_df[train_df["type"] == "Dessert"]
entrees_df = train_df[train_df["type"] == "Entrée"]

print("Entrées :" , sorted(tokenize_ingredients_recettes(entrees_df, nlp).items(), key=lambda x: x[1], reverse=True))
print("Plats : " , sorted(tokenize_ingredients_recettes(plats_df, nlp).items(), key=lambda x: x[1], reverse=True))
print("Desserts :" , sorted(tokenize_ingredients_recettes(desserts_df, nlp).items(), key=lambda x: x[1], reverse=True))
```

Nous remarquons que les mots liés au sucre sont plus dans les recettes de dessert, de même que certains ustensiles comme des moules. Cependant, les entrées et les plats semblent avoir beaucoup d'ingrédients en commun, souvent salés.

On peut souvent distinguer les desserts dans les titres de recettes, la différence entre plat et entrée semble souvent moins facile à détecter. 
Avoir les ingrédients en plus des instructions nous semble utile car les doses permettraient peut-être de différencier les classes, une entrée nécéssite souvent moins d'ingrédients qu'un plat par exemple.
